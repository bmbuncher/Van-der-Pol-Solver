```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from matplotlib import cm
import matplotlib.colors as Colors
import matplotlib.path as path
import matplotlib.lines as mlines
import matplotlib.patches as mpatches

import VDP_funcs as VDP

plt.rc('text', usetex = True)
params = {'text.latex.preamble' : [r'\usepackage{amsmath}']}
plt.rcParams.update(params)

dpi = 360
```

```python
t_max = 5000.0
N_arrows = 10

arrow_locs = np.logspace(-5.0, 0.0, N_arrows)[: -1]
arrow_locs = np.delete(arrow_locs, 1)

figp, axp = VDP.plot_phase_space(t_max = t_max, z0 = 0.1, arrow_locs = arrow_locs, arrowsize = 3, arrow_lw = 0.5, arrow_fc = 'k', arrow_ec = 'k', arrow_style = '-|>', line_color = 'k', cbar_params = [0.92, 0.04, True], norm_type = 'log')

axp.set_title(r'Van der Pol Phase Portrait for $a^+ = 0.1$', fontsize = 20, pad = 10)
figp.get_children()[2].set_title(r'$\lvert\vec{v}\rvert$', fontsize = 20, pad = 10)

zp_text = axp.text(0.51, 0.99, r'$\boldsymbol{z^+}$', fontsize = 22.0, bbox = {'facecolor': 'w', 'edgecolor': 'w', 'alpha': 0.9}, ha = 'left', va = 'top', transform = axp.transAxes)

zm_text = axp.text(0.51, 0.02, r'$\boldsymbol{z^-}$', fontsize = 22.0, bbox = {'facecolor': 'w', 'edgecolor': 'w', 'alpha': 0.9}, ha = 'left', va = 'bottom', transform = axp.transAxes)

VDP.plot_F(axp)

figp.canvas.draw()

figp.savefig('./ppt_figs/zvx_x0_2-5_z0_0-1_static.png', format = 'png', bbox_inches = 'tight', pad_inches = 0.02, dpi = dpi)
```
